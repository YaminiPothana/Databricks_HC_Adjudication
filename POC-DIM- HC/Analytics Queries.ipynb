```python
%sql
SELECT 
count(distinct(claim_id)) AS total_claims
FROM healthcare_adjudication.healthcare_adjudication_gold.fact_claims;
```

```python
%sql
SELECT 
COUNT(DISTINCT patient_id) AS total_patients
FROM healthcare_adjudication.healthcare_adjudication_gold.dim_patient;
```

```python
%sql
SELECT 
COUNT(claim_id) AS approved_claims
FROM healthcare_adjudication.healthcare_adjudication_gold.fact_claims
WHERE status = 'Approved';
```

```python
%sql
SELECT 
p.specialty,
SUM(f.claim_amount) AS total_claim_amount
FROM healthcare_adjudication.healthcare_adjudication_gold.fact_claims f
JOIN healthcare_adjudication.healthcare_adjudication_gold.dim_provider p
ON f.provider_id = p.provider_id
GROUP BY p.specialty
ORDER BY total_claim_amount DESC;
```

```python
%sql
SELECT
COUNT(CASE WHEN status = 'Approved' THEN 1 END) * 1.0 
/ COUNT(*) AS approval_ratio
FROM healthcare_adjudication.healthcare_adjudication_gold.fact_claims;
```

```python
%sql
SELECT
SUM(claim_amount) AS approved_claim_amount
FROM healthcare_adjudication.healthcare_adjudication_gold.fact_claims
WHERE status = 'Approved';
```

```python
%sql
SELECT 
po.plan,
COUNT(f.claim_id) AS total_claims
FROM healthcare_adjudication.healthcare_adjudication_gold.fact_claims f
JOIN healthcare_adjudication.healthcare_adjudication_gold.dim_policy po
ON f.policy_id = po.policy_id
GROUP BY po.plan
ORDER BY total_claims DESC;
```
